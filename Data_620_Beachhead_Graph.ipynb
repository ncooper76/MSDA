To explore the features of networkx, I will now make a few graphs using different methods.

Now I will make an edge list bases on the tiny.net data.

```python
tiny_edge = ["1 2 1",
             "1 3 1",
             "1 4 1",
             "1 5 1",
             "1 6 1",
             "1 7 1",
             "1 8 1",
             "1 9 1",
             "1 10 1",
             "2 3 1",
             "2 4 1",
             "2 5 1",
             "2 6 1",
             "2 7 1",
             "2 9 1",
             "3 4 1",
             "3 5 1",
             "4 5 1",
             "4 6 1",
             "6 7 1",
             "6 9 1",
             "6 10 1"]
h = net.parse_edgelist(tiny_edge, nodetype = int,data=(('weight',int),))
plt.subplot(3,1,3)
net.draw(h, with_labels = True)
plt.show()
```
